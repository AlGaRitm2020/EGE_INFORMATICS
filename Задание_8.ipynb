# Тип 1: Стандартная формулировка

(№ 4447) (А. Богданов) МАРИНА из букв своего имени составляет слова перестановкой исходных букв. Сколько различных слов может составить МАРИНА, если первая буква не может быть гласной?

```python
def isnt_sound(i):
  return (i != 'А' and i != "И" )


from itertools import product

li = product("МАРИНА", repeat=6)



new = []
for word in li:
  word = "".join(word)
  if set(word) == set("МАРИНА") and word.count("А") == 2 and isnt_sound(word[0]):

    new.append(word)
print(len(set(new)))
```

## Тип 2: Слово в отсортированной последовательности по индексу

(№ 2790) (А.Н. Носкин) Петя составляет список из 4-буквенных слов, в состав которых входят только буквы О, С, Е, Н, Ь. Петя расположил слова в обратном алфавитном порядке. Вот начало списка:
1. ЬЬЬЬ
2. ЬЬЬС
3. ЬЬЬО
4. ЬЬЬН
5. ЬЬЬЕ
6. ЬЬСЬ
...

Запишите слово, которое стоит в этом списке под номером 100.

```python



from itertools import product

li = ["".join(i) for i in product("ЬСОНЕ", repeat=4)]

print(li[99])
```

# Тип 3: Не содержать подстроку
(№ 4256) (А. Куканова) Оля составляет 5-буквенные слова из букв К, У, С, А, Т, Ь, причём слова не должны начинаться на мягкий знак и содержать сочетание СУК. Буквы в слове не должны повторяться. Сколько различных слов может составить Оля?

```python
from itertools import permutations

print(len(["".join(i) for i in permutations("КУСАТЬ", r=5) if ("".join(i)[0] !="Ь" and "СУК" not in "".join(i))]))
```
