## 1 тип: перестановка исходных букв

(№ 4447) (А. Богданов) МАРИНА из букв своего имени составляет слова перестановкой исходных букв. Сколько различных слов может составить МАРИНА, если первая буква не может быть гласной?

```python
def isnt_sound(i):
  return (i != 'А' and i != "И" )


from itertools import product

li = product("МАРИНА", repeat=6)



new = []
for word in li:
  word = "".join(word)
  if set(word) == set("МАРИНА") and word.count("А") == 2 and isnt_sound(word[0]):

    new.append(word)
print(len(set(new)))
```

## Тип 2: составление из цифр с условием

(№ 4254) (А. Куканова) Полина составляет 5-значные числа в 5-ичной системе счисления, которые содержат не более 3 чётных цифр. Сколько различных чисел может составить Полина?

```python
def is_less_than_3_even(n):
  evens = len([i for i in n if int(i) % 2 == 0])
  return evens <= 3


from itertools import product

li = product("01234", repeat=5)



new = []
for word in li:
  word = "".join(word)
  if is_less_than_3_even(word) and word[0] != '0':
    new.append(word)
print(new)
print(len(set(new)))
```
