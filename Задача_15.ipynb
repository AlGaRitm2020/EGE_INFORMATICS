(№ 4170) (Е. Джобс) Обозначим через ДЕЛ(n, m) утверждение «натуральное число n делится без остатка на натуральное число m». Сколько существует натуральных значений A, при которых формула
ДЕЛ(A, 5) ∧ (¬ДЕЛ(2020, A) → (ДЕЛ(x, 1718) → ДЕЛ(2023, A)))

тождественно истинна (то есть принимает значение 1 при любом натуральном значении переменной х)?

```python
def DEL(n, m):
  return n % m == 0
result = []
for a in range(1, 10000):
  all = True
  for x in range(1, 10000):
    if not (DEL(a, 5) and not DEL(2020,a)) <= (not DEL(x, 1718)) <= DEL(2023, a):
      all = False
  if all == True:
    result.append(a)
print(len(set(result)), set(result))
```
